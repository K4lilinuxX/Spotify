```python
import requests
import time

# Definimos la API Key de Last.fm (ya tengo creada mi cuenta)
API_KEY = "d638622c0f55ebf99d24d8febd412851"
url_base = "http://ws.audioscrobbler.com/2.0/"
diccionario_tags = {}

# Obtenemos los artistas únicos desde mi DataFrame 'df' ya existente en memoria
artistas_unicos = df['Artist'].unique()

print(f"Iniciando extracción para {len(artistas_unicos)} artistas...")

for artista in artistas_unicos:
    parametros = {
        'method': 'artist.gettoptags',
        'artist': artista,
        'api_key': API_KEY,
        'format': 'json',
        'autocorrect': 1
    }
    
    try:
        respuesta = requests.get(url_base, params=parametros)
        datos = respuesta.json()
        
        if 'toptags' in datos and 'tag' in datos['toptags']:
            # Tomamos los 5 tags más relevantes asignados por la comunidad
            lista_tags = [tag['name'] for tag in datos['toptags']['tag'][:5]]
            diccionario_tags[artista] = ", ".join(lista_tags).lower()
        else:
            diccionario_tags[artista] = "no encontrado"
            
    except Exception as e:
        diccionario_tags[artista] = "error"
    
    # Pausa de 200ms para mantener las peticiones seguras y estables
    time.sleep(0.2)

print("Extracción completada!")

# Mapeamos las etiquetas directamente a la nueva columna del DataFrame
df['Generos'] = df['Artist'].map(diccionario_tags)

# Creamos las máscaras booleanas puras para la clasificación (evitando loc/iloc)
es_latino = df['Generos'].str.contains('latin|reggaeton|mexicano|urbano|español|ranchera|corrido|salsa|bachata', case=False, na=False)
es_anglo = df['Generos'].str.contains('pop|rap|hip hop|rock|country|uk|english|indie', case=False, na=False) & ~es_latino

# Usamos np.select para asignar los grupos
df['Grupo'] = np.select([es_latino, es_anglo], ['Latino', 'Anglosajón'], default='Otro')

# Filtramos el set de datos final para remover los que no entran en la comparación directa
df_analisis = df[df['Grupo'] != 'Otro'].copy()

# Verificamos el resultado en consola
print("\nMuestra de la clasificación final:")
print(df_analisis[['Artist', 'Generos', 'Grupo']].head(10))
```

```python
# ==============================================================================
# ANÁLISIS ESTADÍSTICO Y VISUALIZACIÓN SELECCIONADA (ESTILO PROCEDIMENTAL DIRECTO)
# ==============================================================================

# 1. Configuración de la variable numérica objetivo 
columna_numerica = 'Spotify Popularity'

print("=== INICIANDO PROCESAMIENTO Y ANÁLISIS DE DATOS ===\n")

# 2. Extracción de la distribución de frecuencias por grupo cultural
conteo_grupos = df_analisis['Grupo'].value_counts()
print("1. DISTRIBUCIÓN DE ARTISTAS POR GRUPO CULTURAL:")
print(conteo_grupos)
print("-" * 60)

# 3. Análisis estadístico descriptivo mediante agregación vectorizada
resumen_estadistico = df_analisis.groupby('Grupo')[columna_numerica].agg(
    Total_Artistas='count',
    Promedio_Popularidad='mean',
    Mediana_Popularidad='median',
    Desviacion_Estandar='std',
    Popularidad_Maxima='max',
    Popularidad_Minima='min'
).round(2)

print(f"2. MÉTRICAS DESCRIPTIVAS PARA LA COLUMNA '{columna_numerica}':")
print(resumen_estadistico)
print("-" * 60)


# ==============================================================================
# GRÁFICA 1: COMPARATIVA DEL PROMEDIO DE POPULARIDAD (BARPLOT)
# ==============================================================================
print("Generando Gráfica 1: Promedios...")

# Crear primer lienzo independiente
plt.figure(figsize=(9, 6))

# Pintar gráfico de barras directo
sns.barplot(
    data=df_analisis, 
    x='Grupo', 
    y=columna_numerica, 
    estimator=np.mean, 
    palette="coolwarm",
    errorbar=None
)

# Configuración de etiquetas generales del lienzo
plt.title(f'Promedio de {columna_numerica} por Bloque Cultural', fontsize=13, fontweight='bold', pad=15)
plt.xlabel('Grupo Cultural', fontsize=11)
plt.ylabel(f'Media de {columna_numerica}', fontsize=11)

# Capturar el eje actual para meter las etiquetas de los valores sobre las barras
ax_actual = plt.gca()
for barra in ax_actual.patches:
    altura = barra.get_height()
    ax_actual.annotate(f'{altura:.2f}',
                       xy=(barra.get_x() + barra.get_width() / 2, altura),
                       xytext=(0, 8),
                       textcoords="offset points",
                       ha='center', va='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()  # Se despliega y se limpia la memoria del primer lienzo


# ==============================================================================
# GRÁFICA 2: ANÁLISIS DE DISPERSIÓN Y OUTLIERS (BOXPLOT)
# ==============================================================================
print("\nGenerando Gráfica 2: Dispersión...")

# Crear segundo lienzo completamente independiente
plt.figure(figsize=(9, 6))

# Pintar diagrama de caja directo
sns.boxplot(
    data=df_analisis, 
    x='Grupo', 
    y=columna_numerica, 
    palette="coolwarm",
    width=0.4
)

# Configuración de etiquetas generales del segundo lienzo
plt.title(f'Dispersión y Outliers de {columna_numerica} por Bloque', fontsize=13, fontweight='bold', pad=15)
plt.xlabel('Grupo Cultural', fontsize=11)
plt.ylabel(columna_numerica, fontsize=11)

plt.tight_layout()
plt.show()  # Se despliega la segunda gráfica de forma secuencial

print("\n=== PROCESO FINALIZADO: Gráficas generadas una por una ===")
```
