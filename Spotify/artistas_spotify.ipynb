```python
import requests
import time

# Definimos la API Key de Last.fm (ya tengo creada mi cuenta)
API_KEY = "d638622c0f55ebf99d24d8febd412851"
url_base = "http://ws.audioscrobbler.com/2.0/"
diccionario_tags = {}

# Obtenemos los artistas únicos desde mi DataFrame 'df' ya existente en memoria
artistas_unicos = df['Artist'].unique()

print(f"Iniciando extracción para {len(artistas_unicos)} artistas...")

for artista in artistas_unicos:
    parametros = {
        'method': 'artist.gettoptags',
        'artist': artista,
        'api_key': API_KEY,
        'format': 'json',
        'autocorrect': 1
    }
    
    try:
        respuesta = requests.get(url_base, params=parametros)
        datos = respuesta.json()
        
        if 'toptags' in datos and 'tag' in datos['toptags']:
            # Tomamos los 5 tags más relevantes asignados por la comunidad
            lista_tags = [tag['name'] for tag in datos['toptags']['tag'][:5]]
            diccionario_tags[artista] = ", ".join(lista_tags).lower()
        else:
            diccionario_tags[artista] = "no encontrado"
            
    except Exception as e:
        diccionario_tags[artista] = "error"
    
    # Pausa de 200ms para mantener las peticiones seguras y estables
    time.sleep(0.2)

print("Extracción completada!")

# Mapeamos las etiquetas directamente a la nueva columna del DataFrame
df['Generos'] = df['Artist'].map(diccionario_tags)

# Creamos las máscaras booleanas puras para la clasificación (evitando loc/iloc)
es_latino = df['Generos'].str.contains('latin|reggaeton|mexicano|urbano|español|ranchera|corrido|salsa|bachata', case=False, na=False)
es_anglo = df['Generos'].str.contains('pop|rap|hip hop|rock|country|uk|english|indie', case=False, na=False) & ~es_latino

# Usamos np.select para asignar los grupos
df['Grupo'] = np.select([es_latino, es_anglo], ['Latino', 'Anglosajón'], default='Otro')

# Filtramos el set de datos final para remover los que no entran en la comparación directa
df_analisis = df[df['Grupo'] != 'Otro'].copy()

# Verificamos el resultado en consola
print("Muestra de la clasificación final:")
print(df_analisis[['Artist', 'Generos', 'Grupo']].head(10))
```

## ¿Cuál es la cantidad de artistas anglosajones y latinos que hay?

Agregamos el `drop_duplicates(subset=['Artist', 'Grupo'])`. Esto le dice a Pandas: "Si ves a Eminem 10 veces en la columna 'Artist', quédate solo con la primera y borra las otras 9".

```python
# Usamos el df_analisis para que únicamente salgan los artistas que la API clasificó como Anglosajón y Latino
# Creamos un DataFrame sin duplicados, quedándonos solo con un registro por artista
df_artistas_unicos = df_analisis.drop_duplicates(subset=['Artist', 'Grupo'])

# Contamos la cantidad real de artistas por grupo
grupo_artistas = df_artistas_unicos["Grupo"].value_counts()

# Graficamos
plt.figure(figsize=(12, 6))
plt.title("Cantidad de artistas anglosajones y latinos")
barras = sns.barplot(
    x=grupo_artistas.index,
    y=grupo_artistas.values,
    palette="Blues"
)
# Agregamos el número de artistas anglosajones y latinos que hay 
for contenedores in barras.containers:
    barras.bar_label(contenedores,
                     fmt='%.0f')

plt.xlabel("Grupo del artista")
plt.ylabel("Cantidad de artistas")
plt.grid(axis="y",
         linestyle="--",
         alpha=0.6)
plt.show()
```

## ¿Cuál es el grupo de artistas que tiene mas popularidad en Spotify?

```python
# Agrupamos la popularidad en Spotify por grupo
popularidad_grupos = df_analisis.groupby("Grupo")["Spotify Popularity"].mean()

# Graficamos
plt.figure(figsize=(12,6))
plt.title("Popularidad en Spotify por grupo de artistas")
barras = sns.barplot(x=popularidad_grupos.index,
                     y=popularidad_grupos.values,
                     hue=popularidad_grupos.index,
                     palette="magma")
# Agregamos el promedio de popularidad en Spotify por cada grupo (Anglosajones y Latinos)
for contenedores in barras.containers:
    barras.bar_label(contenedores,
                     fmt='%.2f')
plt.xlabel("Grupo del artista")
plt.ylabel("Promedio de popularidad en Spotify")
plt.grid(axis="y",
         linestyle="-",
         alpha=0.6)
plt.show()
```
