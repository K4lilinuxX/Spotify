# ¿Cómo clasifiqué a los artistas en `Anglosajón` y `Latino`?
## ¿Cómo clasifiqué a los artistas por género musical? (El rol de la API)
La clasificación por género musical no se realizó de manera manual ni con criterios subjetivos; en su lugar, se automatizó el consumo de metadatos globales conectando el script a la API de Last.fm.  

* Consumo de Etiquetas Colaborativas (Folksonomía): Last.fm no utiliza una base de datos de géneros estática, sino que recopila las etiquetas (`tags`) que millones de oyentes en todo el mundo le asignan a los artistas de forma histórica.

* El Proceso Técnico: A través del método `artist.gettoptags` y enviando la firma de nuestra `API_KEY`, el script consultó la plataforma para cada uno de los 2,000 artistas únicos del dataset.

* Filtrado de Relevancia: Para evitar ruido en la información, el código se programó para extraer únicamente las 5 etiquetas más populares y con mayor peso dentro de la respuesta en formato JSON. Estas etiquetas se unieron en una cadena de texto en minúsculas y se almacenaron en una nueva columna llamada `Generos`.

**Ejemplo de salida de la API**: > `Kendrick Lamar` → _"hip-hop, rap, west coast, hip hop, compton"_

## ¿Cómo clasifiqué a los artistas por Anglosajones y Latinos? (El rol del Script)
Es fundamental precisar que la API de Last.fm no divide a los artistas por bloques culturales. La clasificación en los grupos 'Latino' y 'Anglosajón' fue ejecutada al 100% por nuestro código de Python a través de la ingeniería de características (feature engineering).  

Se programaron **máscaras booleanas** vectorizadas y se utilizó la función `np.select()` para evaluar las cadenas de texto de la columna `Generos` bajo las siguientes reglas de negocio estrictas:

## `Latino`
El script escanea las etiquetas de cada artista buscando palabras clave asociadas a la cultura hispanohablante, música regional e industria urbana de Latinoamérica y España. Si los géneros contienen alguno de los siguientes términos, el artista es asignado de forma prioritaria a este bloque:  
**Palabras clave: latin, reggaeton, mexicano, urbano, español, ranchera, corrido, salsa, bachata**  

## `Anglosajón`
Para garantizar la pureza estadística de este grupo y evitar falsos positivos (por ejemplo, un artista latino de reggaetón que también tuviera la etiqueta global `pop`), se implementó una regla de doble validación condicional:

0. Inclusión: El artista debe poseer etiquetas fuertemente vinculadas a la música de habla inglesa o corrientes globales del mainstream: pop, rap, hip hop, rock, country, uk, english, indie.

1. Exclusión Estricta (~es_latino): El artista no debe contener ninguna de las palabras clave del bloque latino. De esta manera, si el algoritmo detecta la palabra pop pero también encuentra la palabra reggaeton, el operador lógico excluye inmediatamente al artista del bloque anglosajón, asegurando que este grupo contenga únicamente música de raíz e influencia netamente angloparlante.

## `Otro`
Cualquier artista cuyas etiquetas no coincidieran con ninguno de los patrones anteriores —como exponentes de K-Pop, música electrónica instrumental pura, o casos donde la API devolvió `"no encontrado"` — fue clasificado bajo la etiqueta neutral de `'Otro'`.

```python
import requests # Usamos requests para comunicarlos con la API (http)
import time

# Definimos la API Key de Last.fm (ya tengo creada mi cuenta)
API_KEY = "d638622c0f55ebf99d24d8febd412851"
url_base = "http://ws.audioscrobbler.com/2.0/"
diccionario_tags = {}

# Obtenemos los artistas únicos desde mi DataFrame 'df' ya existente en memoria
artistas_unicos = df['Artist'].unique()

print(f"Iniciando extracción para {len(artistas_unicos)} artistas...")

for artista in artistas_unicos:
    parametros = {
        'method': 'artist.gettoptags',
        'artist': artista,
        'api_key': API_KEY,
        'format': 'json',
        'autocorrect': 1
    }
    
    try:
        respuesta = requests.get(url_base, params=parametros)
        datos = respuesta.json()
        
        if 'toptags' in datos and 'tag' in datos['toptags']:
            # Tomamos los 5 tags más relevantes asignados por la comunidad
            lista_tags = [tag['name'] for tag in datos['toptags']['tag'][:5]]
            diccionario_tags[artista] = ", ".join(lista_tags).lower()
        else:
            diccionario_tags[artista] = "no encontrado"
            
    except Exception as e:
        diccionario_tags[artista] = "error"
    
    # Pausa de 200ms para mantener las peticiones seguras y estables
    time.sleep(0.2)

print("Extracción completada!")

# Mapeamos las etiquetas directamente a la nueva columna del DataFrame
df['Generos'] = df['Artist'].map(diccionario_tags)

# Creamos las máscaras booleanas puras para la clasificación (evitando loc/iloc)
es_latino = df['Generos'].str.contains('latin|reggaeton|mexicano|urbano|español|ranchera|corrido|salsa|bachata', case=False, na=False)
es_anglo = df['Generos'].str.contains('pop|rap|hip hop|rock|country|uk|english|indie', case=False, na=False) & ~es_latino

# Usamos np.select para asignar los grupos
df['Grupo'] = np.select([es_latino, es_anglo], ['Latino', 'Anglosajón'], default='Otro')

# Filtramos el set de datos final para remover los que no entran en la comparación directa
df_analisis = df[df['Grupo'] != 'Otro'].copy()

# Verificamos el resultado en consola
print("Muestra de la clasificación final:")
print(df_analisis[['Artist', 'Generos', 'Grupo']].head(10))
```
